```python
import pandas as pd
import os

# redirect to current notebook directory
NOTEBOOK_DIR = os.getcwd()

# build correct file path
csv_path = os.path.join(NOTEBOOK_DIR, "SGJobData.csv", "SGJobData.csv")

# load csv
df = pd.read_csv(csv_path)

# preview
df.head()
```

```python
#sieve out all invalid values
df.shape
df.columns
df.info()
```

```python
#drop unneeded categories
df = df.drop(columns= ["occupationId", "metadata_isPostedOnBehalf", "status_id", "metadata_jobPostId"], errors="ignore")
df.info()
```

```python
#for each category, return % of invalid values
import numpy as np
import pandas as pd

def invalid_percentage(df):
    results = []
    
    for col in df.columns:
        total = len(df)
        
        # Count invalids
        invalid = (
            df[col].isna() |
            (df[col].astype(str).str.strip() == "")
        ).sum()
        
        percent = (invalid / total) * 100
        
        results.append([col, invalid, round(percent, 4)])
    
    return pd.DataFrame(
        results, 
        columns=["column", "invalid_count", "invalid_percentage"]
    ).sort_values(by="invalid_percentage", ascending=False)

invalid_stats = invalid_percentage(df)
invalid_stats
```

Since invalid values are 0.3803% for each affected category, just drop invalid data.

```python
df = df.dropna()
for col in df.columns:
    df = df[df[col].astype(str).str.strip() != ""]
    
invalid_stats = invalid_percentage(df)
invalid_stats
```
